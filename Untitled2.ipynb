```python
import random

def get_hint(name, guess):
    hint = ""
    for i in range(len(name)):
        if name[i] == guess[i]:
            hint += name[i]
        else:
            hint += "_"
    return hint

def main():
    print("Welcome to the Name Guessing Game!")
    print("Player 1, please think of a name.")
    name = input("Player 1, enter the name you're thinking of: ").lower()

    # Clear the console to hide the name from Player 2
    print("\n" * 50)

    print("Player 2, it's your turn to guess the name!")
    attempts = 5
    while attempts > 0:
        guess = input(f"Attempt {6-attempts}: Enter your guess: ").lower()

        if guess == name:
            print("Congratulations! You guessed the name correctly.")
            break
        else:
            attempts -= 1
            hint = get_hint(name, guess)
            print(f"Wrong guess. Here's a hint: {hint}")

    if attempts == 0:
        print(f"Sorry, you've run out of attempts. The name was: {name}")

if __name__ == "__main__":
    main()
```
